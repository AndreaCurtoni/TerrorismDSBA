```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
gtds = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="ISO-8859-1")
pd.set_option("display.max_rows", None, "display.max_columns", None)
missing_value = gtds.isna().sum()
gtdsDropped = gtds
for i in range(0,missing_value.shape[0]-1,1):
    if ((missing_value.values[i]/len(gtds))>0.80):
        gtdsDropped = gtdsDropped.drop(columns=[missing_value.index[i]])
gtdsDropped.shape
```

```python
gtdsDropped.columns
```

Capire gruppi terroristici pi√π interessati ai riscatti
Capire gruppi terroristici e i loro obiettivi (crit1,crti2,crti3)

```python
ds = gtdsDropped[['gname','crit1','crit2','crit3']]
print(len(ds))
ds.head(5)
```

```python
ds = ds.groupby(['gname']).agg('sum')
print(len(ds))
ds.sort_values(by='crit1', ascending=False).head(10)
```
